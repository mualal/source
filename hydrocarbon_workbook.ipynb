## Режимы работы скважины в пласте

Соотношение воронки депрессии скважины к границам области дренирования.

Упражнение (длительность неустановившегося режима)

```python
C_t = 3E-4
phi = 0.2
mu = 3
r_b = 1000
k = 100E-3
kappa = k/(phi*mu*C_t)
h = 10

T = 35*r_b**2/kappa
print('Длительность неустановившегося режима', T/(60*60*24), 'суток')
```

Скважину запускают не сразу. Как правило, её выводят на режим.

Пробурили, спуск ГИС, интерпретация, спуск эксплуатационной колонны, убедились в герметичности (опрессовка), проперфорировали, пригнали флот ГРП, спуск ГНКТ (гибкие насосно-компрессорные трубы) с пакером, ГРП, промывка пропанта из скважины, глушение скважины, бригада КРС спускает погружной ЭЦН (крепится к ГНКТ), при спуске насоса ГНКТ накручиваются, спустили насос (скважина не пустая, в ней есть жидкость глушения, чтобы не допустить газонефтепроявления на устье), есть кабель питания для ЭЦН.

Далее кнопочный пуск (начинает работать ЭЦН). С этого момента начинается вывод скважины на режим. Частота насоса увеличивается с низких до рабочих частот. Дебит жидкости постепенно растёт, а обводнённость падает (постепенно вытягивается вся жидкость глушения). Процесс от кнопочного пуска до вывода параметров по скважине (как по жидкости, так и по обводнённости) называют выводом скважины на режим. 7 суток - среднее время вывода скважины на режим

## Режимы работы пласта

Отражают энергию, за счёт которой нефть движется к забою скважины.

При водонапорном режиме снижение давления компенсируется притоком воды из аквифера.

Гравитационный режим

SAGD = парогравитационный дренаж. Пласт битума, в него забуривают горизонтальную скважину: ближе к подошве добывающая скважина. На расстоянии 5-10 м сверху от неё параллельно бурится паронагнетательная скважина. Вырабатывается расширяющаяся паровая камера. И битум стекает по краям этой паровой камеры в добывающую скважину.


## Формула Дюпюи и обобщения (формулы Паркера)

Общая формула производительности скважины на псевдоустановившемся режиме в пласте сложной формы при постоянном дебите на скважине.

Форм-факторы (классические форм-факторы получены при отсутствии перетока на границе пласта и постоянном дебите на скважине)

Значения классического форм-фактора практически не зависят от формы пласта, если скважина находится в центре и размеры пласта по длине и ширине практически равны (пласт не вытянут). Если же пласт вытянут или скважина сильно смещена из центра, то форм-фактор существенно зависит от формы пласта.

Если на скважине задано постоянное давление, то форм-факторы отличаются от классических.

Производительность скважины при смешанных условиях на границе пласта. Метод источников для расчёта форм-факторов.

Пунктир = постоянное давление на границе пласта (ППД), организуется расстановкой нагнетательных и добывающих скважин в шахматном порядке.

Сплошная линия = условие неперетока, организуется расстановкой только добывающих скважин.

Упражнения (по форм-факторам)


```python
import numpy as np

phi = 0.2
mu = 10
k = 100E-3
p_e = 250  # среднее
p_w = 50
h = 10
B = 1.2
r_e = 250
r_w = 0.108
S = -4

Q = k*h*(p_e-p_w)/(18.4*mu*B*(np.log(r_e/r_w)+S-3/4))
print('Производительность по формуле Дюпюи', Q, 'куб.м/сут')
```

```python
r_e = 50
Q = k*h*(p_e-p_w)/(18.4*mu*B*(np.log(r_e/r_w)+S-3/4))
print('Производительность по формуле Дюпюи', Q, 'куб.м/сут')
Q = k*h*(p_e-p_w)/(18.4*mu*B*(1/2*np.log(4*200*800/(np.e**(0.57722)*0.115*r_w**2))+S))
print('Производительность по формуле с форм-фактором', Q, 'куб.м/сут')
```
